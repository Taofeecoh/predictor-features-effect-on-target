### Feature Engineering
#### Accordiing to the data dictionary above, there are 3 Borrowing Entities:
- Central Government
- State Owned Enterprise
- Joint Ventures

```python
# Delimiter function
def delimiter(value):
    if isinstance(value, str):
        if '|' in value:
            value = value.split('|')[1]
        else:
            value    
    return value
```

```python
df['BorrowerType2'] = df['BorrowerType'].apply(delimiter)
df.head()
```

```python
print(df['BorrowerType2'].unique())
```

```python
print(len(df['BorrowerType2'].unique()))
```

```python
patterns = [r'State', r'mental', r'Private', 'Oth', 'Agency']

# Recategorize 
for p in patterns:
    df.loc[df['BorrowerType2'].str.contains(p, regex=True, na=False), 'BorrowerType2']= 'Government Agency/SOE'

df.loc[df['BorrowerType2'].str.contains(' bank', regex=True, na=False), 'BorrowerType2']= 'Central government'
```

```python
print('Unique Borrower Types (Re-categorized): ', len(df['BorrowerType2'].unique()))
print('\n',df['BorrowerType2'].unique().tolist())
```

```python
df.info()
```

```python
# Fill missing values with the most frequent value RecategorizedBorrowerType for this class of Creditor Group.

Multilateral_CreditorGroup_Mode = df.loc[(df['CreditorGroup'] == 'Multilateral') & (df['BorrowerType2'].notna())]['BorrowerType2'].mode()

df.loc[
    (df['CreditorGroup'] == 'Multilateral') & 
    (df['BorrowerType2'].isna()), 
    'BorrowerType2'
] = Multilateral_CreditorGroup_Mode[0]
```

```python
# Fill missing values with the most frequent value in RecategorizedBorrowerType for this class of Creditor Group.

Bilateral_CreditorGroup_Mode = df.loc[(df['CreditorGroup'] == 'Bilateral') & (df['BorrowerType2'].notna())]['BorrowerType2'].mode()

df.loc[
    (df['CreditorGroup'] == 'Bilateral') & 
    (df['BorrowerType2'].isna()), 
    'BorrowerType2'
] = Bilateral_CreditorGroup_Mode[0]
```

```python
# Fill missing values with the most frequent value in RecategorizedBorrowerType for this class of Creditor Group.

Private_CreditorGroup_Mode = df.loc[(df['CreditorGroup'] == 'Private') & (df['BorrowerType2'].notna())]['BorrowerType2'].mode()

df.loc[
    (df['CreditorGroup'] == 'Private') & 
    (df['BorrowerType2'].isna()), 
    'BorrowerType2'
] = Private_CreditorGroup_Mode[0]
```

```python
df.info()
```

### Slice data

```python
df1 = df[['interest', 'maturity', 'grace', 'BorrowerType2','CreditorGroup','Amount_musd']].copy()
df1.info()
```

```python
# Drop missing ponts from subsets
df1 = df1.dropna(subset=['Amount_musd', 'CreditorGroup', 'BorrowerType2'])
```

### Regression analysis with statsmodels.api : Model 1

```python
import statsmodels.api as sm

# Split into target and feature variables
X = data.drop(['Amount_musd'], axis=1)
y = data['Amount_musd']

# Set constant term for the intercept
X1 = sm.add_constant(X)
model1 = sm.OLS(y,X1).fit()

# Print the summary of the model
print(model1.summary())
```

## Investigating Multicolinearity with Variance Inflation Factor (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif['variables'] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)

X_vif = X.copy()
calc_vif(X_vif)
```

```python
# Dropping correlated predictive features to reduce multicolinearity

X.drop(['CreditorGroup_Bilateral'], axis=1, inplace=True)
```

```python
X_vif = X.copy()
calc_vif(X_vif)
```

### Model 2

## Report Summary

#### Significance of Variables
Statistically, a Variable is considered significant if its p-value is less than 5% (0.05) and its confidence interval is greater than 95%. 
From this model, the significant variables with their corresponding p-values are listed below:
interest (0.0001)
maturity (0.0001)
grace (0.0001)
CreditorGroup_Multilateral (0.0001)
CreditorGroup_Private (0.0001)

From the values above, it is statistically safe to conclude that these variables are statistically significant with p < 0.05 generally.

#### Model Fit
The R-squared value of 0.167 indicates that approximately 16.7% of the variance in the amount of debt taken by African countries is explained by the independent variables in the model even though the majority of the independent variables are statistically significant.
Summarily, a 16.7% score is suggestive of a relatively low goodness-of-fit, translating to further improvement of the model.

#### Impact Assessment
Variable CreditorGroup_Private has the greatest impact on the amount of debt, with the highest coefficient at 159.9364. This could be translated as loans from private creditors having a close relationship with a significantly higher amount of debt compared to the baseline. 
A phenomenon that could occur because private creditors often offer larger loans, with different terms and conditions as compared to other creditors.

#### F-statistic
The F-statistic of 211.8, and its associated p-value is very close to zero (9.20e-288). The F-statistic tests whether at least one of the predictor variables is significantly related to the dependent variable. The very low p-value indicates that the model is statistically significant overall, meaning that at least one of the predictors is significantly related to the amount of debt.

#### Multicollinearity
The VIF table above gives an insight into the Multicolinearity level in the model1, with VIF infinity values suggestive of Structural or Data-related Multicolinearity.

Potential effects of multicollinearity include:
- Fluctuation in estimation of coefficients due to its instability
- Difficulty interpreting the individual effect of the correlated predictive features on the target.
